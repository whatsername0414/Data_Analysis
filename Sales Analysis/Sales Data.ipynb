```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
```

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
```

```python
files = [file for file in os.listdir("./Sales_Data")]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("all_data.csv", index=False)
```

```python
all_data = pd.read_csv("all_data.csv")
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
all_data = all_data.dropna(how='all')
all_data
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

## What was the best month for sales? How much was earned that month?

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):    
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

```python
month_results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
```

```python
months = range (1, 13)

plt.bar(months, month_results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month')
plt.show()
```

## What city had the highest number of sales?

```python
city_results = all_data.groupby('City').sum()
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, city_results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')
plt.show()
```

## What time should display advertisements to maximize likelyhood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.ylabel('Number of Orders')
plt.xlabel('Hour')
plt.grid()
plt.show
```

So the time to display ads are around 10am to 11am or 6pm to 7pm because those are the peaks on where the orders in the data sets are happened.

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(10)
```

```python
df['Group'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df.head()
```

```python
df = df[['Order ID', 'Group']].drop_duplicates()
df.head()
```

## What product are most sold often together?

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Group']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```

Above are the items that are most sold together. By this we make a promotion like when buying iPhone we can offer a smart deal with charging cable.

## What product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
products = [product for product, df in product_group]
plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical', size=8)
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.grid()
plt.show()
```

Batteries, Charging Cable, Wired Headphone are the most sold product. In my opinion, this items sold the most because they are cheap.

```python
prices = all_data .groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)
plt.show()
```

It is evident that when the quantity ordered is high the price is low. We can see it on the items Batteries, Charging Cable, Wired Headphone. This items have low price so people are more willing to pay that price.
